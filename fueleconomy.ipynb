```python
import pandas as pd
import numpy as np
from sqlalchemy import create_engine
```

```python
df = pd.read_csv("fuel_econ.csv")
df.head()
```

```python
fuelecon = df[['id', 'make', 'model', 'year', 'VClass', 'drive', 'fuelType', 
               'feScore']].copy()
```

```python
fuelecon.head()
```

```python
fuelecon = fuelecon.rename(columns={
    'id': 'ID',
    'make': 'Make',
    'model': 'Model',
    'year': 'Year',
    'VClass': 'Vehicle_Class',
    'drive': 'Drive',
    'fuelType': 'Fuel_Type',
    'feScore': 'Fuel_Economy_Score'
})
```

```python
fuelecon['Good Fuel Economy?'] = np.where(fuelecon['Fuel_Economy_Score'] > 6, 'yes', 'no')
```

```python
fuelecon.set_index('ID', inplace = True)
```

```python
fuelecon.sample(5)
```

```python
fuelecon[fuelecon["Model"]== "Mustang Convertible"]
```

```python
fuelecon = fuelecon.drop_duplicates(subset=['Model', 'Year'])
```

```python
fuelecon[fuelecon["Model"]== "Mustang Convertible"]
```

```python
rds_connection_string = "postgres:postgres@localhost:5432/norwaycars"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
engine.table_names()
```

```python
fuelecon.to_sql(name='fueleconomy', con=engine, if_exists='replace', index=True)
```

```python
with engine.connect() as con:
    con.execute('ALTER TABLE fueleconomy ADD PRIMARY KEY ("Model", "Year");')
```
